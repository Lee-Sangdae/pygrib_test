```python
import pygrib
import numpy as np

def read_grib_file(grib_file_path, target_latitude, target_longitude, parameter_name):
    # GRIB 파일 열기
    grbs = pygrib.open(grib_file_path)

    # 해당 변수명과 위경도에 맞는 GRIB 메시지 찾기
    target_grbs = grbs.select(name=parameter_name, latitude=target_latitude, longitude=target_longitude)

    if not target_grbs:
        raise ValueError(f"No matching GRIB message found for parameter '{parameter_name}' at the specified latitude and longitude.")

    # 가장 가까운 GRIB 메시지 선택
    closest_grb = min(target_grbs, key=lambda grb: grb.distance(target_latitude, target_longitude))

    # 데이터 읽기
    data = closest_grb.values

    # 기타 GRIB 메시지 속성 읽기
    level = 'surface'#closest_grb.level
    valid_time = closest_grb.validDate

    # GRIB 파일 닫기
    grbs.close()

    return data, level, valid_time

# 사용 예시
grib_file_path = 'g120_v070_erea_unis_h009.2023122518.gb2'
target_latitude = 37.0  # 예시 위도
target_longitude = 127.0  # 예시 경도
parameter_name = 'Temperature'  # 예시 변수명

data, level, valid_time = read_grib_file(grib_file_path, target_latitude, target_longitude, parameter_name)

# 여기서 data에는 해당 위치의 온도 데이터가, level에는 해당 데이터의 수직 레벨, valid_time에는 해당 예보의 시각이 저장됩니다.
```
