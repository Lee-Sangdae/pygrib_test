```python
import pygrib

def extract_grib_values(grib_file_path, short_names):
    # Open the GRIB2 file
    grbs = pygrib.open(grib_file_path)

    # Dictionary to store extracted values for each short name
    extracted_values = {}

    for short_name in short_names:
        # Loop through all messages in the GRIB2 file
        for grb in grbs:
            # Check if the current message has the desired short name
            if grb.shortName == short_name:
                # Extract the values and add them to the dictionary
                values = grb.values
                extracted_values[short_name] = values

    # Close the GRIB2 file
    grbs.close()

    return extracted_values

# Specify the path to your GRIB2 file
grib_file_path = "hres_0.1deg_unis_h012_2024010100.grib2"

# Specify the short names of the variables you want to extract (e.g., 10u, 10v)
short_names_to_extract = ["10u", "10v"]

# Extract values for the specified variables
result = extract_grib_values(grib_file_path, short_names_to_extract)

# Print the extracted values
for short_name, values in result.items():
    print(f"Variable: {short_name}")
    print("Values:", values)
    print("=" * 50)
```

```python
import pygrib
from pyproj import Proj
import numpy as np

def extract_grib_values(grib_file_path, target_lat, target_lon):
    # Open the GRIB2 file
    grbs = pygrib.open(grib_file_path)

    # Dictionary to store extracted values for each variable
    extracted_values = {}

    for grb in grbs:
        # Extract latitude, longitude, and values
        lats, lons = grb.latlons()
        values = grb.values

        # Find the closest grid point to the target latitude and longitude
        lat_idx = np.abs(lats[:, 0] - target_lat).argmin()
        lon_idx = np.abs(lons[0, :] - target_lon).argmin()

        # Get the variable name
        short_name = grb.shortName

        # Extract the value at the target location
        target_value = values[lat_idx, lon_idx]

        # Add the value to the dictionary
        extracted_values[short_name] = target_value

    # Close the GRIB2 file
    grbs.close()

    return extracted_values

# Specify the path to your GRIB2 file
grib_file_path = "hres_0.1deg_unis_h012_2024010100.grib2"

# Specify the target latitude and longitude
target_latitude = 37.5  # Example latitude
target_longitude = 127.0  # Example longitude

# Extract values for the specified variables at the target location
result = extract_grib_values(grib_file_path, target_latitude, target_longitude)

# Print the extracted values
for short_name, value in result.items():
    print(f"Variable: {short_name}")
    print(f"Value at ({target_latitude}, {target_longitude}): {value}")
    print("=" * 50)
```

```python
import pygrib
from pyproj import Proj
import numpy as np

def extract_grib_values(grib_file_path, target_lat, target_lon, target_variables):
    # Open the GRIB2 file
    grbs = pygrib.open(grib_file_path)

    # Dictionary to store extracted values for specified variables
    extracted_values = {}

    for grb in grbs:
        # Extract latitude, longitude, and values
        lats, lons = grb.latlons()
        values = grb.values

        # Find the closest grid point to the target latitude and longitude
        lat_idx = np.abs(lats[:, 0] - target_lat).argmin()
        lon_idx = np.abs(lons[0, :] - target_lon).argmin()

        # Get the variable name
        short_name = grb.shortName

        # Check if the variable is in the target_variables list
        if short_name in target_variables:
            # Extract the value at the target location
            target_value = values[lat_idx, lon_idx]

            # Add the value to the dictionary
            extracted_values[short_name] = target_value

    # Close the GRIB2 file
    grbs.close()

    return extracted_values

# Specify the path to your GRIB2 file
grib_file_path = "hres_0.1deg_unis_h012_2024010100.grib2"

# Specify the target latitude and longitude
target_latitude = 37.5  # Example latitude
target_longitude = 127.0  # Example longitude

# Specify the target variables
target_variables = ["10u", "10v"]

# Extract values for the specified variables at the target location
result = extract_grib_values(grib_file_path, target_latitude, target_longitude, target_variables)

# Print the extracted values
for short_name, value in result.items():
    print(f"Variable: {short_name}")
    print(f"Value at ({target_latitude}, {target_longitude}): {value}")
    print("=" * 50)
```
